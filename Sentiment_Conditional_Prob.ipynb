```python
def posnegCheck(sentence):

  #sentence = input('Enter Sentence: ')

  poss=0
  negs=0
  words = sentence.split()

  for word in words:
    if pdf['pos'].eq(word).any():
      poss+=1
      #print("Positive word", word)
    elif ndf['neg'].eq(word).any():
      negs+=1
      #print("Negative word", word)

  score = (poss-negs)
  #print(score)
  print(sentence)
  if (score > 0):
    print("Sentence is positive")
  elif (score < 0):
    print("Sentece is negative")
  else:
    print("Sentence is neutral")

  conf = max(poss,negs)/len(sentence)
  print("Confidence: ", conf)

posnegCheck("NLP is cool")
posnegCheck("NLP is cool and useful")
posnegCheck("NLP is hard")
posnegCheck("NLP is hard and useless")
posnegCheck("NLP stand for natural language processing")
```

```python
import numpy as np
import csv
import re
import math
from collections import Counter

def naiveBayesClassifier(trainingset, testset):

  # Blank list of all and IE class words
  megaDocGB = ''
  megaDocIE = ''

  probIEint=0
  probGBint=0
  probNeut=0

  # Casefolding both lists
  train = [list(map(str.casefold, x)) for x in trainingset]
  #print(train)

  #this doesnt work for splitting sentence into individual words easily
  #test = [list(map(str.casefold, x)) for x in testset]
  #print(test)

  testCase = []

  for test in testset:
    testCase.append((test[0]).casefold())

  # Writing megadocs into blank class lists for IE and GB
  # Also interates an integer for calculating probability of GB or IE class based on test


  for i in range(len(train)):
    if train[i][1] == 'gb':
      tempGB = train[i][0]
      megaDocGB = megaDocGB + " " + tempGB
      probGBint+=1
    elif train[i][1] == 'ie':
      tempIE = train[i][0]
      megaDocIE = megaDocIE + " " +  tempIE
      probIEint+=1
    else:
      probNeut+=1


  probTot=probIEint+probGBint
  probGB = (probGBint/probTot)
  probIE = (probIEint/probTot)

  print("megaDocGB=", megaDocGB)
  print("megaDocIE=", megaDocIE)

  print("ProbGB=", probGB)
  print("ProbIE=", probIE)

  tokenGB = megaDocGB.split()
  tokenIE = megaDocIE.split()

  vocab = megaDocGB + megaDocIE
  V = vocab.split()

  GBcount = Counter(tokenGB)
  IEcount = Counter(tokenIE)

  Vcount = Counter(V)
  #vTot = sum(Vcount.values())
  vTot = len(Vcount)

  GBcount = dict(GBcount)
  IEcount = dict(IEcount)
  Vcount = dict(Vcount)

  GBnum = len(GBcount)
  IEnum = len(IEcount)

  print("GB_BOW = ", GBcount)
  print("IE_BOW = ", IEcount)
  print("V = ", Vcount)

  print("|V| = ", vTot)

  wordConditionalProbGB = 0
  wordConditionalProbIE = 0

  for test in testCase:
    print("Test document = ", test)

    probGBdoc = probGB
    probIEdoc = probIE

    for word in test.split():
      if word in GBcount:
        wordConditionalProbGB = (GBcount[word] + 1) / (GBnum + vTot)
      else:
        wordConditionalProbGB = (0 + 1) / (GBnum + vTot)
      if word in IEcount:
        wordConditionalProbIE = (IEcount[word] + 1) / (IEnum + vTot)
      else:
        wordCondtionalProbIE = (0 + 1) / (IEnum + vTot)

      print(' word = ', word, '   wordConditionalProbGB = ',wordConditionalProbGB , '   wordConditionalProbIE = ', wordConditionalProbIE)

      probGBdoc = probGBdoc * wordConditionalProbGB
      probIEdoc = probIEdoc * wordConditionalProbIE

      print('probGBdoc= ', probGBdoc, ', probIEdoc= ', probIEdoc)

      if (math.isclose(probGBdoc, probIEdoc)):
        print('Class inferred (close) = IE/GB')

      elif probIEdoc > probGBdoc:
        print('Class inferred = IE')
      elif probGBdoc > probIEdoc:
        print('Class inferred = GB')

      print('============================================================================')

  return

trainingset = [('London is the capital of GB', 'GB'),('Oxford is a city in GB', 'GB'),('Dublin is the capital of Ireland', 'IE'),('Limerick is a city in Ireland', 'IE')]
testset = [('University of Limerick', '?'), ('University College Dublin', '?'), ('Imperial College London','?'), ('University of Oxford','?'), ('Ireland & GB', '?')]

naiveBayesClassifier(trainingset, testset)
```
