```python
# Takes input of a nonword, with any spelling
# Provides list of most likely intended word/spelling based on word count corpus

import jellyfish

def nonWordSpellingCorrection(nonWord):


  comp = unigrams_df[1].sum()

  cWords = [str(word) for word in unigrams_df.index.tolist() if jellyfish.damerau_levenshtein_distance(nonWord, str(word)) <= 1]

  cWordsFreqs = unigrams_df[1][cWords].tolist()

  cWordsProbs = [cWord / comp for cWord in cWordsFreqs]

  cWordsDict = {'Word': cWords, 'Frequency': cWordsFreqs, 'P(word)': cWordsProbs}
  correctWordDF = pd.DataFrame(cWordsDict)

  return correctWordDF


nonWordSpellingCorrection("Acces")
```
