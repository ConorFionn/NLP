Standard Probability of Sentences:

a = "I want to want"\
b = "I want to spend"

$P(a) = \frac{P(I, want)}{P(want)} * \frac{P(want, to)}{P(to)} *\frac{P(to, want)}{P(want)}$


$a = \frac{827}{927} \frac{608}{927} * \frac{0}{927} = 0 $

$a' = 1 - a = 1 $


$P(b) = \frac{P(I, want)}{P(want)} * \frac{P(want, to)}{P(to)} *\frac{P(to, spend)}{P(spend)}$

$b = \frac{827}{927} \frac{608}{927} * \frac{211}{278} = 0.17$

$b' = 1 - b = 0.83 $

With Add-One Laplace Smoothing:

$P(a) = \frac{P(I, want)+V}{P(want)+V} * \frac{P(want, to)+V}{P(to)+V} *\frac{P(to, want)+V}{P(want)+V}$


$a = \frac{827+V}{927+V} \frac{608+V}{927+V} * \frac{0+V}{927+V} = 0 $

$a' = 1 - a = 1 $


$P(b) = \frac{P(I, want)+V}{P(want)+V} * \frac{P(want, to)+V}{P(to)+V} *\frac{P(to, spend)+V}{P(spend)+V}$

$b = \frac{827+V}{927+V} \frac{608+V}{927+V} * \frac{211+V}{278+V} = 0.17$

$b' = 1 - b = 0.83 $

```python
import pandas as pd

!wget https://norvig.com/ngrams/count_1w.txt
filePath = "/content/count_1w.txt"

!wget https://norvig.com/ngrams/count_2w.txt
filePath2 = "/content/count_2w.txt"
unigrams = open(filePath, 'r')
bigrams = open(filePath2, 'r')

unigrams_df = pd.read_table(filePath, header=None)
bigrams_df = pd.read_table(filePath2, header=None)
unigrams_df.columns = ['word','count']
bigrams_df.columns = ['word','count']

totalUnigrams = len(unigrams_df)

totalBigrams = len(bigrams_df)

print(f'Number of unigrams: {totalUnigrams} Number of Bigrams: {totalBigrams}')

display(unigrams_df.head(100),bigrams_df.head(100))
```

```python
def probability(sentence):

  words = sentence.split()
  chance = []

  for i in range(len(words)-1):
    probW = []
    wordC = words[i]
    wordC_next = words[i+1]

    for k in range(totalUnigrams):
      test = unigrams_df.iloc[k,0]
      spl = test.split()

      if spl[0] == wordC:
        probW.append(spl[1])
        break

    for k in range(totalBigrams):
      test = bigrams_df.iloc[k,0]
      spl = test.split()

      if spl[0] == wordC:

        if spl[1] == wordC:
          probW.append(spl[2])
          break

    print(f'unigram count for "{wordC}" = {probW[0]}' )
    print(f'bigram count for "{wordC} {wordC_next}" = {probW[1]}')
    first = probW[0]
    first = int(first)
    second = probW[1]
    second = int(second)
    probablity = second+1/first+i
    print(f'bigram probablity for "{wordC} {wordC_next}" = {probablity}')
    print('----------------------------------------------------\n')
    chance.append(probablity)
    probW.clear()


  sentenceProbability  = chance[0] * chance[1]

  return sentenceProbability


probability("i love you")>probability('i hate you')

```
